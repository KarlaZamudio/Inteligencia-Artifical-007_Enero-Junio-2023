2.Desarrolla un algoritmo que por medio de una función,le pide al usuario una lista de N elementos y como resultados invertir el sentido de la lista.

```python
def invertir_lista():
  x = int(input("Escribe la cantidad de datos que tendra tu la lista: "))
  lista=[]
  for i in range(x):
    dato = input("Ingresa el dato de tu lista : ")
    lista.append(dato)
  lista_invertida = list(reversed(lista))
  return lista_invertida
lista_invertida = invertir_lista()
print("La lista invertida queda de la siguiente manera: ", lista_invertida)
```
